```python
import pickle
import os
from tqdm.auto import tqdm
```

# All Language Preprocessing for CNN

```python
languages = ['english_0','english_1','filipino','chinese','korean']
```

```python
os.mkdir('comparative-abusive-lang/data/cnn/')
```

```python
for lang in tqdm(languages):
    proc_data = pickle.load(open(f'processed_data/{lang}_processed.pkl','rb'))

    my_dict = {'text_data':[{'train':proc_data['text_train'],'valid':proc_data['text_test'],'test':proc_data['text_test']}]*10,
               'context_data':[{'train':['']*len(proc_data['text_train']),'valid':['']*len(proc_data['text_test']),'test':['']*len(proc_data['text_test'])}]*10,
               'label_data':[{'train':proc_data['label_train'],'valid':proc_data['label_test'],'test':proc_data['label_test']}]*10
              }
    with open(f'comparative-abusive-lang/data/cnn/{lang}_cnn.pkl','wb') as f:
        pickle.dump(my_dict,f)
        
```
